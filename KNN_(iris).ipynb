```python
from numpy import genfromtxt
data_path = '/content/gdrive/MyDrive/Talukder AI/iris.csv'
my_data = genfromtxt(data_path, delimiter=',')
print(type(my_data))
```

```python

data_iris = my_data.tolist()
print(data_iris)

print(type(data_iris))
```

#Dataset preparation

```python
import random
random.shuffle(data_iris)
print(data_iris)
print(len(data_iris))
```

```python


Train_set=[]
Val_set=[]
Test_set=[]

for s in data_iris:
  R = random.uniform(0, 1)
  if R>=0 and R<=0.7:
    Train_set.append(s)
  elif R>0.7 and R<=0.85:
    Val_set.append(s)
  else:
    Test_set.append(s)


print(len(Train_set))
print(len(Val_set))
print(len(Test_set))
print(len(Train_set) + len(Test_set) + len(Val_set))
```

```python
#KNN Classification for Val_set(Use Iris data):

from scipy.spatial import distance
import operator
k = 7

correctly_classified = 0

for v in Val_set:
  max_class = {}
  L = {}
  for t in Train_set:
    dis = distance.euclidean(v[0:(len(v)-1)], t[0:(len(t)-1)])
    L[dis] = t
  
  sort_new_L = sorted(L.keys())
  counter = 1

  for j in sort_new_L:
    if int(L[j][-1]) in max_class.keys(): 
      max_class[int(L[j][-1])] = max_class[int(L[j][-1])]+1
    else:
      max_class[int(L[j][-1])] = 0
      max_class[int(L[j][-1])] = max_class[int(L[j][-1])]+1
    counter = counter + 1
    if(counter > k):
      break
  vali = max(max_class.items(), key=operator.itemgetter(1))[0]

  if(int(v[-1]) == vali):
    correctly_classified = correctly_classified + 1
  
print('Accuracy:', (correctly_classified/ len(Val_set)) * 100, '%')
```
